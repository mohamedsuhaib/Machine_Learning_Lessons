##### Demo: Feature Reduction
__Problem Statement:__ The dataset you are going to use in this practice is the famous Iris data set. The 
dataset consists of 150 records of Iris plant with four features: “sepal-length”, “sepal-width”, “petal-
length”, and “petal-width”. All the features are numeric. The records have been classified into one of the 
three classes, that is, “Iris-setosa”, “Iris-versicolor”, or “Iris-verginica”.

__Objective:__
* Train the models on original number of features
* Reduce the number of variables by merging correlated variables
* Extract the most important features from the dataset that are responsible for maximum variance in 
the output.

```python
#Step 1: import the Libraries and Dataset
import sklearn
from sklearn.datasets import load_iris
data = load_iris()

#Step 2: Define the data and target values
X=data.data
Y=data.target

#Step 3: Split the data and target values into training and testing stes
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test = train_test_split(X,Y, test_size=0.2,random_state=1)
print(X_train.shape)
```

```python
#Step 4: Implement a Logistic Regression Model
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression(solver='liblinear')
```

```python
lr.fit(X_train,Y_train)
#Since we are building a model and there are all probable outcomes.
#we must compare the same with actual ones. so us to determine predictive power of our model using accuracy score
Y_predict = lr.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy = accuracy_score(Y_predict,Y_test)
print(accuracy)
```

We got 83% accuracy. We may get better accuracy, if we implement the above model with more relevant features. Let's try Post PCA Feature Reduction and compare the results. It will reduce the no.of component originally present in our data to a certain number such that it explains 95% of the variants.

```python
#Step 5: Perform PCS transformation
#Post PCA Feature Reduction and compare the results
from sklearn.decomposition import PCA
sklearn_PCA = PCA(n_components=0.95)
sklearn_PCA.fit(X_train)
X_train_transformed = sklearn_PCA.transform(X_train)
print(X_train_transformed.shape)
print(X_test.shape)
```

```python
X_test_transformed = sklearn_PCA.transform(X_test)
print(X_test_transformed)
```

From the output, the no.of features had reduced considerably retainng 95% of the variants.

Now we have transformed our features using PCA, let's implement the model once again over the transformed dataset and check the results.

```python
print(X_test_transformed.shape)
```

```python
#Step 6: Implement Logistics Regression over transformed dataset
lr = LogisticRegression(solver='liblinear',penalty='l1')
lr.fit(X_train_transformed,Y_train)
Y_predict = lr.predict(X_test_transformed)

#Step 7: Check for Model accuracy
from sklearn.metrics import accuracy_score
accuracy = accuracy_score(Y_predict,Y_test)
print(accuracy)
```

The model accuracy went up from 83% to 96%.

##### Demo: Labeled Feature Reduction
__Problem Statement:__ Consider the iris dataset preloaded within the mlxtend library. Segregate the 
data and labels accordingly and transform the data to two linear discriminants. 

__Objective:__ Perform LDA as a dimensionality reduction technique.

```python
#Import Libraries and dataset
from mlxtend.data import iris_data
from mlxtend.preprocessing import standardize
from mlxtend.feature_extraction import LinearDiscriminantAnalysis

#Define data and target value
X,y = iris_data()
X = standardize(X)

#Instantiate LDA
lda=LinearDiscriminantAnalysis(n_discriminants=2)
lda.fit(X,y)

#Transform the X variable using LDA
X_lda = lda.transform(X)
X_lda
```

Thus, we have transformed our data using LDA.

__Objective:__ Understand and practice LDA using Scikit learn.

###### Import and Split
Import the load_digits dataset from scikit learn and split it accordingly

```python
import sklearn
from sklearn.datasets import  load_digits
digits = load_digits()
X = digits.data
y = digits.target
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, 
random_state=1)
print(X_train.shape)
```

###### Number of Components in Transformed Shape
Fit LDA on the training data

```python
from sklearn.linear_model import LogisticRegression
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis as lda
sklearn_lda = lda()
sklearn_lda.fit(X_train,y_train)
X_train = sklearn_lda.transform(X_train)
print("Number of components in transformed shape {}".format(X_train.shape[1]))
```

###### Dataset Transformation
Transforming the test data

```python
#sklearn_lda.transform(X_test)
X_test =sklearn_lda.transform(X_test)
print(X_test.shape)
```

###### Fitting a Logistic Regression Model

```python
lr = LogisticRegression(solver='liblinear',penalty='l1')
lr.fit(X_train,y_train)
y_predict = lr.predict(X_test)
from sklearn.metrics import accuracy_score
accuracy = accuracy_score(y_predict,y_test)
print(accuracy)
```
