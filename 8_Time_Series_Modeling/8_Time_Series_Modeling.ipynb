## Demo: Air Passengers - I
__Problem Statement:__ The Air Passenger dataset provides monthly total of US airline passengers, from 1949 to 1960. The objective of this analysis and modeling is to review time series theory.

__Objective:__

We will be following an ARIMA modeling procedure of the AirPassengers dataset as follows:
* Perform the Exploratory Data Analysis
* Deomposition of data
* Check for the stationarity of your data using Rolling Statistics and Dickey fuller test
* If stationarity is present, remove it using differencing in Python.

```python
import numpy as np
import pandas as pd
from datetime import datetime as dt
from statsmodels.tsa.stattools import adfuller, acf, pacf
from statsmodels.tsa.arima_model import ARIMA
import math
import matplotlib.pyplot as plt
%matplotlib inline
from matplotlib.pylab import rcParams
rcParams['figure.figsize']= 15,6
import warnings
warnings.filterwarnings('ignore')
```

```python
#Load the dataset
data = pd.read_csv('AirPassengers.csv')

#The dataset consists of data and monthly passenger count values
data['Month'].head()
```

```python
data['Month'] = data['Month'].apply(lambda X: dt(int(X[:4]),int(X[5:]),15))
data = data.set_index('Month')
data.head()
```

```python
#assigning the passenger data time series variable
ts = data['#Passengers']

#Plotting the time Series
plt.plot(ts)
```

It is clear from the plot, there is a overall increase in trend with some seasonality in it.

```python
#Transforming series using Log decomposition
ts_log = np.log(ts)
```

```python
#Stationarity Check
#Function for performing Rolling Statistics and Dickey-Fuller test
def test_stationarity(timeseries):
    rolmean = timeseries.rolling(window=52, center=False).mean()
    rolstd = timeseries.rolling(window=52, center=False).std()
    orig = plt.plot(timeseries, color='blue', label='original')
    mean = plt.plot(rolmean, color='red', label='Rolling Mean')
    std = plt.plot(rolstd, color='black', label='Rolling Std')
    plt.legend(loc='best')
    plt.title('Rolling Mean and Standard Deviation')
    plt.show(block = False)
    print('Results of Dickey-Fuller Test:')
    #initializing df test variable for oerforming Dickey-Fuller Test
    dftest = adfuller(timeseries, autolag='AIC')
    #Initializing o/p witht he test statistics, Pvalue and other required variables
    dfoutput = pd.Series(dftest[0:4], index=['Test Statistics','PValue','#lags Used', 'No. of Observation used'])
    #Initializing the critical value variable
    for key, value in dftest[4].items():
        dfoutput['Critical value (%s)'%key]= value
    print(dfoutput)
    
#lets apply the data to a function and see the results for stationarity.
test_stationarity(data['#Passengers'])
```

By observing the plot, we infer the test statistics is more than critical and moving average is not constant over time. So, the Null-hypothesis of Dickey-Fuller test cannot be rejected. This shows that the time series is not stationary.

```python
#Decomposing the data and plot the moving average
plt.plot(ts_log)
```

```python
#initializing the moving average variable
movingAverage = ts_log.rolling(window=12).mean()

#initializing the moving Standard Deviation variable
movingSTD = ts_log.rolling(window=12).std()

#plotting the data
plt.plot(ts_log)

#plotting the Moving Average
plt.plot(movingAverage, color='red')
```

```python
ts_log_mv_diff = ts_log - movingAverage
ts_log_mv_diff.head(12)
```

```python
ts_log_mv_diff.dropna(inplace=True)
ts_log_mv_diff.head(10)
```

```python
#parse our time series data into this function and check for stationary
test_stationarity(ts_log_mv_diff)
```

The results of Dickey-fuller test shows that the test statistics is less than 10% of critical value. This tells we are 99% confident that time series is stationary. we can reject the Null-hypothesis.

## Practice Project: Beer Prodcution - I
__Problem Statement:__ The Beer production dataset provides a time series data for monthly beer production in Australia, for the period Jan 1956  Aug 1995.

__Objective:__
 Check for the stationarity of your data using Rolling Statistics and Dickey fuller test
 if stationarity is present, remove it using differencing in Python

```python
#Step 1: Data Import
import numpy as np
import pandas as pd
import math
import matplotlib.pyplot as plt
from matplotlib.pyplot import rcParams
from datetime import datetime
%matplotlib inline
df1 = pd.read_csv('monthly-beer-production-in-austr.csv')
df1.head()
```

```python
#Step 2: Parse and Plot
dateparse1 = lambda dates: pd.datetime.strptime(dates, '%Y-%m')
data1 = pd.read_csv('monthly-beer-production-in-austr.csv', 
                    parse_dates=['Month'], index_col='Month', date_parser=dateparse1)
ts1 = data1['Monthly beer production in Australia']
ts1.plot()
plt.ylabel("Consumption in megalitres")
```

```python
#Step 3: Stationarity Check
from statsmodels.tsa.stattools import adfuller
def test_stationarity1(timeseries1):
    #Determing rolling statistics
    rolmean1 = timeseries1.rolling(window=52,center=False).mean() 
    rolstd1 = timeseries1.rolling(window=52,center=False).std()
    #Plot rolling statistics:
    orig1 = plt.plot(timeseries1, color='blue',label='Original')
    mean1 = plt.plot(rolmean1, color='red', label='Rolling Mean')
    std1 = plt.plot(rolstd1, color='black', label = 'Rolling Std')
    plt.legend(loc='best')
    plt.title('Rolling Mean & Standard Deviation')
    plt.show(block=False)
    #Perform Dickey-Fuller test:
    print ('Results of Dickey-Fuller Test:')
    dftest1 = adfuller(timeseries1, autolag='AIC')
    dfoutput1 = pd.Series(dftest1[0:4], index=['Test Statistic','p-value',
                                             '#Lags Used','Number of Observations Used'])
    for key,value in dftest1[4].items():
        dfoutput1['Critical Value (%s)'%key] = value
    print (dfoutput1)

test_stationarity1(data1['Monthly beer production in Australia'])
```

The test statistic is more than critical value and the moving average is not constant over time. So, the null hypothesis of the Dickey-Fuller test cannot be rejected. This shows that the time series is not stationary.

```python
#Step 4: Stationarize
ts_log_mv_diff1 = data1['Monthly beer production in Australia'].apply(lambda x: math.log(x)).rolling(2).mean().diff(1)
ts_log_mv_diff1.dropna(inplace=True)
#ts_log_mv_diff.plot() 
test_stationarity1(ts_log_mv_diff1)
```

Test statistic < 5 % of critical value. Rejectnull hypothesis

## Demo: Air Passengers - II
__Problem Statement:__ The Air Passenger dataset provides monthly total of US airline passengers, from 1949 to 1960. This dataset is of a time series class.

__Objective:__
* Perform ARIMA modeling in Python after obtaining ACF and PACF plots
* Calculate the forecasts

__Forecasting a Time Series__
Model the Stationary time series(already completed code), Use statistical modeling technique called ARIMA to forecast the data where there are dependencies in the values.

To find out the p and q values for the model, we will use ACF and PACF plots.

```python
#Initializing and plotting ACF
plt.plot(np.arange(0,11), acf(ts_log_mv_diff, nlags=10))
plt.axhline(y=0,linestyle='--', color='gray')
plt.axhline(y=-7.96/np.sqrt(len(ts_log_mv_diff)),linestyle='--', color='gray')
plt.title("Autocorrelation function")
plt.show()                            
```

The AFC curve crosses the upper confidence value when the lagged value is b/w 0 and 1. Thus the optimal value of q in the ARIMA model must be '0' or '1'.

```python
#Plotting PACF (Initialize plot and applying acf)
plt.plot(np.arange(0,11),pacf(ts_log_mv_diff,nlags=10))
plt.axhline(y=0,linestyle='--',color='gray')
plt.axhline(y=-7.96/np.sqrt(len(ts_log_mv_diff)),linestyle='--',color='gray')
plt.axhline(y=7.96/np.sqrt(len(ts_log_mv_diff)),linestyle='--',color='gray')
plt.title('Partial Autocorrelation Function')
plt.show()
```

The PACF curve drops zero b/w lagged values 1 and 2. Thus the optimal value of p in the ARIMA model is 1 or 2.5.

```python
#Step 4: Fit a model using an ARIMA algorithm and calculate the forecast
model = ARIMA(ts_log, order=(1,1,0))
results_ARIMA = model.fit(disp=-1)
plt.plot(ts_log_mv_diff)
plt.plot(results_ARIMA.fittedvalues,color='red')
plt.title('RSS: %.4f'%sum((results_ARIMA.fittedvalues[1:] - ts_log_mv_diff)**2))
```

```python
#model predictions
#Store the predicted value results as a seperate series and observe it.
predictioms_ARIMA_diff = pd.Series(results_ARIMA.fittedvalues, copy=True)
predictioms_ARIMA_diff.head()
```

These dates strats from 1949-02-15, not the first month. this is because we took a lag by 1 and the first element doesn't have anything before to subtract from. The way to convert the differencing to log scale is to add these differences consequently to the base number.

A easier way to do is to first determine the cumulative sum of index and then add it to the base number.

The cumulative sun can be found as,

```python
predictioms_ARIMA_diff_cumsum = predictioms_ARIMA_diff.cumsum()
predictioms_ARIMA_diff_cumsum.head()
```

Doing some calculation using the previous o/p to check these are correct, next we add them to the base number. For this, create a series with all values as the base number and add the difference to it.

```python
predictioms_ARIMA_log = pd.Series(ts_log.iloc[0], index=ts_log.index)
predictioms_ARIMA_log = predictioms_ARIMA_log.add(predictioms_ARIMA_diff_cumsum,fill_value=0)
predictioms_ARIMA_log.head()
```

```python
#initializing ARIMA model
predictioms_ARIMA = np.exp(predictioms_ARIMA_log)
plt.plot(ts)
plt.plot(predictioms_ARIMA)
plt.title('RSE: %.4f'%np.sqrt(sum((predictioms_ARIMA - ts)**2)/len(ts)))
```

The red curve is our prediction which has RMSE 52.1841, while our model is good at predicting the trend, it fails in capturing the seasonality.
