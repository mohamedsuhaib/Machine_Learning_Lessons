# Supervised Learning
Supervised Learning is a type of machine learning used to train models from labeled training data. It allows you to predict output for future or unseen data.


* Labelled Data is a group of samples or features that are tagged with one or more labels. 
* Training data needs to be fed in ML algorithm in order for it to generate a ML Model that learns from our algorithm.
* The learning algorithm discovers patterns in the training data such that the input parameter corresponds to the target.
* The output of the training process is the Machine Learning Model,which we can use for predictions.

## Types of Supervised Learning
In supervised learning, algorithm is selected based on target variable.

### Classification Algorithm
If target variable is categorical (classes), then use classification algorithm.
In other words, classification is applied when the output has finite and discreet values.

Example: Predict the class of car given its features like horsepower,mileage, weight, colour, etc.

The classifier will build its attributes based on these features. Analysis has three potential outcomes - Sedan, SUV, or Hatchback

Classification predicts categorical class labels or classifies information that is used in the classification of characteristics to build a model based on the training set and values. In Classification graph, the red line is the Classifier, and the points above and below are two classes. In each class, the distance from the nearest point to the red line is the "Error of Classification".



### Regression Algorithm
If target variable is a continuous numeric variable (1002000), then use a regression algorithm. In other words, regression analysis is used when we want to predict a continous dependent variable from a number of independent variables.

Example: Predict the price of a house given its sq. area, location, no of bedrooms, etc.    

A simple regression algorithm is given below
* y = w * x + b

This shows relationship between price (y)  and sq. area (x), where price is a number from a defined range.

In Regression graph involves in finding the best fit straight line through the plotted points. The best fit line is called as Regression line(Red Diagonal line). The Regression line consists of the predicted score on 'Y' for each possible value of 'X'. The error of prediction for each point is the distance of that point from the regression line. The point that is very near to the red line has small error of prediction and vice versa for the point that is far from the line.

## Demo: Linear Regression
__Problem Statement:__ The Advertising dataset captures sales revenue generated with respect to advertisement spends across multiple channels like radio, tv, and newspaper. In this demo, we will use linear regression to understand how spending on advertisements impact sales.

Here are some of the questions which can reveal information from the data.

* What types of ads contribute to sales?

__Objective:__

__Build a linear regression model to:__
* Interpret the coefficients of the model
* Make predictions

* Evaluate model efficiency using RMSE and R-Square values

```python
#import the Libreries
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.metrics import r2_score, mean_squared_error
from math import sqrt
%matplotlib inline
```

```python
#Import the dataset in pandas dataframe
data = pd.read_csv('D:\\NIPUN_SC_REC\\3_Practice_Project\\Course_6_Machine Learning\\Lesson_Recap\\4_Supervised_Learning\\Advertising.csv',index_col=0)
#Independent variables are TV, Radio and Newspaper withAmount of dollars spent in thousands.
#Target Variable is Sales of single product in a gicen market with thousands of widgets.
data.columns= ['TV', 'Radio', 'Newspaper', 'Sales']
data.head()
```

```python
data.shape
```

### Step 1: Visualize the relationship between the features and the target variable using scatter plots

```python
#Initializing the plot
fig,axs = plt.subplots(1,3,sharey=True)
data.plot(kind='scatter', x='TV', y='Sales',ax=axs[0],figsize=(16,8))
data.plot(kind='scatter', x='Radio', y='Sales',ax=axs[1])
data.plot(kind='scatter', x='Newspaper', y='Sales',ax=axs[2])
```

We do see a relationship between ads and sales. Applying linear regression analysis to estimate sales and television as spending.

```python
#Lets use TV as feature column
feature_cols=['TV']
#x is the independent variable
x = data[feature_cols]
#y is the dependent variable
y = data.Sales
```

### Step 2: Apply Linear Regression

```python
#Import Linear Regression model
from sklearn.linear_model import LinearRegression

#Initializing the Model
ln = LinearRegression()

#Fitting the model on X and Y
ln.fit(x,y)
```

```python
#Print the intercept and coefficient of the resulting linear equation
print(ln.intercept_)
print(ln.coef_)
```

Let's say that there was a new market where TV advertising spends were $50,000. What would we predict for the sales in that market?. 

Use the previous results to calculate the new values for sales.

```python
7.032593549127695 + 0.04753664 * 50
```

Let's predict the new X values

```python
X_new = pd.DataFrame({'TV':[50]})
X_new.head()
```

Using the above equation, we would now predict the sales of 9,409 widgets in that market.

```python
ln.predict(X_new)
```

Let's make predictions for the smallest and the largest observed values of x. And then use the predicted values to plot the latest squares line.

```python
X_new = pd.DataFrame({'TV':[data.TV.min(),data.TV.max()]})
X_new.head()
```

```python
preds = ln.predict(X_new)
preds
```

```python
#Initializing scatter plot for TV and Sales
data.plot(kind='scatter', x='TV', y='Sales')
#Plotting to least squared line
plt.plot(X_new, preds, c='red', linewidth=2)
```

Looking at the plot, we can say that we have a high bias and a low variance model.

### Step 3: Hypothesis testing and p-values

Use the null-hypothesis, let's assume there is no relationship between tv ads and sales. Now let's test this hypothesis. we shall reject the null-hypothesis, if the 95% confidence interval does not include zero.

```python
#importing stat models
import statsmodels.formula.api as smf

#Applying ols on TV sales
ln=smf.ols(formula='Sales ~ TV',data=data).fit()
```

```python
ln.conf_int()
```

```python
#Printing the p-Values
ln.pvalues
```

From the output, The p-values represnts the possibility that the coefficient is actually Zero. In this case, the p-value for TV is far less than 0.5, and hence we infer that there is a relationship between tv ads and sales.

How well the model fits the data?

The most common way to evaluate the overall fit of a linear model is by using the r-Squared value. Let's print the r-Squared value.

```python
ln.rsquared
```

The threshold for a good r-Squared value depends widely on the domain. Therefore, it's most useful as a tool for comparing different models.

### Step 4: Multiple Linear Regression

Earlier we used TV as the feature column, now let's take all tha independent variables to build a multiple linear regression model.

```python
#Use the three variables to build the model
feature_cols = ['TV', 'Radio', 'Newspaper']
X = data[feature_cols]
y = data.Sales

#import model selection
from sklearn import model_selection
#Creating a train and test
xtrain, xtest, ytrain, ytest = model_selection.train_test_split(X,y,test_size=0.3,random_state=42)

#Applying linear regression
ln = LinearRegression()
ln.fit(X,y)
print(ln.intercept_)
print(ln.coef_)
```

```python
ln = LinearRegression()
ln.fit(xtrain,ytrain)
```

```python
print(ln.intercept_)
print(ln.coef_)

#Printing the Mean-Squared-Error (MSE)
predictions = ln.predict(xtest)
print(sqrt(mean_squared_error(ytest,predictions)))
```

Using the OLS regression results, we get lot of information in the model summary output.

```python
ln = smf.ols(formula='Sales ~ TV + Radio + Newspaper', data=data).fit()
ln.conf_int()
```

```python
ln.summary()
```

Let's understand the key things that we've inferred from this output.
1. TV and radio have significant p-values, whereas newspaper does not. Thus we reject the null-hypothesis for TV and Radio. There is no association between these features and sales. However, we fail to reject the null-hypothesis for the variable newspaper.
2. TV and radio ad spending are both positively associated with sales, whereas newspaper ad spending is slightly negatively associated with sales. However, this is irrelevant since we have failed to reject the null-hypothesis for the variable newspaper.
3. This model has a higher r-squared value '0.897, than the previous model, which means that this model provides a beter fit to the data than a model that only includes the TV variable.

### Step 5: Feature Selection

How do we know what features need to be included in a linear model?

Here's some ideas,
1. Try out different models and only keep the predictors in the model if they have small p-values.
2. Check whether the r-squared value goes up when you add new predictors.

```python
ln = smf.ols(formula='Sales ~ TV + Radio', data=data).fit()
ln.rsquared
```

```python
ln = smf.ols(formula='Sales ~ TV + Radio + Newspaper', data=data).fit()
ln.rsquared
```

Observing the output, we get simillar r-squared values. R-squared will always increase as we add more features to the model. Even if they are unrelated to the response. Thus, selecting the model with the highest R-squared is not a reliable approach for choosing the best linear model.

### Train test split

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3,random_state=1, stratify=y)
print('Labels counts in y:', np.bincount(y))
print('Labels counts in y_train:', np.bincount(y_train))
print('Labels counts in y_test:', np.bincount(y_test))
```

### Standardizing features

```python
sc = StandardScaler()
sc.fit(X_train)
X_train_std = sc.transform(X_train)
X_test_std = sc.transform(X_test)
```

### Plotting decision surface

```python
def plot_decision_regions(X, y, classifier, test_idx=None, resolution=0.02):
    # setup marker generator and color map
    markers = ('s', 'x', 'o', '^', 'v')
    colors = ('red', 'blue', 'lightgreen', 'gray', 'cyan')
    cmap = matplotlib.colors.ListedColormap(colors[:len(np.unique(y))])
    
    # plot the decision surface
    x1_min, x1_max = X[:, 0].min() - 1, X[:, 0].max() + 1
    x2_min, x2_max = X[:, 1].min() - 1, X[:, 1].max() + 1
    xx1, xx2 = np.meshgrid(np.arange(x1_min, x1_max, resolution),np.arange(x2_min, x2_max,resolution))
    Z = classifier.predict(np.array([xx1.ravel(), xx2.ravel()]).T)
    Z = Z.reshape(xx1.shape)
    plt.contourf(xx1, xx2, Z, alpha=0.3, cmap=cmap)
    plt.xlim(xx1.min(), xx1.max())
    plt.ylim(xx2.min(), xx2.max())
    for idx, cl in enumerate(np.unique(y)):
        plt.scatter(x=X[y == cl, 0], y=X[y == cl, 1],alpha=0.8,c=colors[idx],marker=markers[idx], label=cl, edgecolor='black')
    if test_idx:
        X_test, y_test = X[test_idx, :], y[test_idx]
        plt.scatter(X_test[:, 0], X_test[:, 1],c='',edgecolor='black',alpha=1.0,linewidth=1,marker='o',s=100, label='test set')        
X_combined_std = np.vstack((X_train_std, X_test_std))
y_combined = np.hstack((y_train, y_test))        
```

### Training  logistic regression model

```python
lr = LogisticRegression(C=100.0, random_state=1)
lr.fit(X_train_std, y_train)
plot_decision_regions(X_combined_std, y_combined, classifier=lr, test_idx=range(105, 150))
plt.xlabel('petal length [standardized]')
plt.ylabel('petal width [standardized]')
plt.legend(loc='upper left')
plt.tight_layout()
#plt.savefig('images/03_06.png', dpi=300)
plt.show()
```
