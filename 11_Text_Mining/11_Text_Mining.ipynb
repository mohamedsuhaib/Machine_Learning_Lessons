```python
#Lets remove the stop words using the English stop word list in NLTK library
#We will be using 'Set' checking as it is faster in python than a 'List'.
stopword_en = set(stopwords.words('english'))

#By removing all stopword from the text we obtain..
print([word for word in single_tokenized_lowered if word not in stopword_en])
```

__Stemming and Lemmatization__

we will be using stemming and lemmatization to reduce words to their root form.

```python
#Step 3: Perform Stemming and Lemmatizing
from nltk.stem import PorterStemmer
porter = PorterStemmer()

for word in single_tokenized_lowered:
    print(porter.stem(word))
```

```python
from nltk.stem import WordNetLemmatizer
wnl = WordNetLemmatizer()

#Printing the root words
for word in single_tokenized_lowered:
    print(wnl.lemmatize(word))
```

```python
#Step 4: Remove the stop words
#We also need to evaluate the POS target for each token
stop_words = set(stopwords.words('english'))
tokenized = sent_tokenize(data2)

for i in tokenized:
    wordlist = nltk.word_tokenize(i)
    words = [w for w in wordlist if not w in stop_words]
    tagged = nltk.pos_tag(wordlist)
    print(tagged)
```

For our final text processing task we will be applying named entity recognition to classify named entities in text into predefined categories such as the names of persons, organization, locations, expressions of times, quantities, monetary values, percentages etc.,.

Now enter the tagged_sentences in chunk parser, if we set the parametr binary wquals True, then named entitiesare just tagged as 'ne', otherwise the classifier adds category label such as person, organization and Geo-Political Entity(GPE).

```python
#Step 5: Perform Named Entity Recognition
sentences = nltk.sent_tokenize(data2)
tokenized_sentences = [nltk.word_tokenize(sentence) for sentence in sentences]
tagged_sentences = [nltk.pos_tag(sentence) for sentence in tokenized_sentences]
chunked_sentences = nltk.ne_chunk_sents(tagged_sentences, binary=True)
```

```python
#Create a fucntion along with entity names
def extract_entity_name(t):
    entity_names = []
    
    #We will extarct named entities from nltk chunked expression 
    #And store them in the empty list created above
    if hasattr(t,'label') and t.label:
        if t.label() == 'NE':
            entity_names.append(''.join([child[0] for child in t]))
        else:
            for child in t:
                entity_names.extend(extract_entity_name(child))
                
    return entity_names
```

Again we will set the entity names list as a empty list and we will extract the entity names by iterating over each tree and chunk sentences.

```python
entity_names = []
for tree in chunked_sentences:
    entity_names.extend(extract_entity_name(tree))
    
print(set(entity_names))
```

### Chunking Using Python

```python
#Lets consider the sentence below:
import nltk
from nltk import word_tokenize

sent = "The little mouse ate the fresh cheese"

#Using word_tokenize, we can find out the different tags associated with the sentence
sent_tokens = nltk.pos_tag(word_tokenize(sent))
sent_tokens
```

```python
nltk.help.upenn_tagset()
```

### NP Chunk and Parser
We will now create grammar from a noun phrase and will mention the tags we want in our chunk phrase within{}. 

Here we have created a regular expression matching the string.

```python
#The regular expression indicates, optional Determiner followed by Optional number of Adjective followd by a Noun
grammar_np = r"NP: {<DT>?<JJ>*<NN>}"
```

we will now have to parse the chunk. Therefore, we will create a chunk parser and pass our noun phrase string to it.

```python
chunk_parser = nltk.RegexpParser(grammar_np)
```

The parser is now ready. We will use the parse ( ) within our chunk parser to parse our sentence.

```python
chunk_result = chunk_parser.parse(sent_tokens)
chunk_result
```

The sentence provided is "The little mouse ate the fresh cheese". This sentence has been parsed and the tokens that match the regular expressions are chunk together into noun phrases 'NP'.

### VP Chunk and Parser
Create a verb phrase chunk using regular expressions. The regular expression has been defined as a optional personal pronoun followed by 'Zero' or 'more verbs' with any of its type followed by any type of adverb.

```python
grammar_vp = r"vp: {<PRP>?<VB|VBD|VBZ|VBG>*<RB|RBR>?}"
```

Well now create another chunk_parser and pass the verb phrase string to it.

```python
chunk_parser2 = nltk.RegexpParser(grammar_vp)
```

Create another sentence and tokenize it. Add POS tags to it.

```python
sent3 = "She is walking quickly to the mall"
sent_tokens3 = nltk.pos_tag(word_tokenize(sent3))
```

Now, use the new verb phrase parser to parse the tokens and run the results.

```python
chunk_result3 = chunk_parser2.parse(sent_tokens3)
chunk_result3
```

The output diagram shows a verb parser where a pronoun followed by two verbs and an adverb are chunked together into a verb phrase.

## Chinking


### Create Chink Grammar
Consider we create a chinking grammar string containing three things:
* Chunk name
* The regular expression sequence of a chunk
* The regular expression sequence of our chink

Here, In the given code, we have the chunk regular expression as optional personal pronoun followed by 'Zero or' more occurances of any type of the verb type followed bt 'zero or more' occurances of any of the adverb types.

```python
chink_grammar = r"""
chk_name: #chunk name
{<PRP>?<VB|VBD|VBZ|VBG>*<RB|RBR>?} #chunk regex sequence
}<RB>+{ #chink regex sequence - adverb
"""
```


The chink regular expression says that it heads to check for the adverb in the extracted chunk and remove it from the chunk. Inside the chinking block with }{, we have created one or more adverbs.

### Create Chink Parser

We will now create a parser from nltk.RegexpParser and pass the chink_grammar to it.

```python
chink_parser = nltk.RegexpParser(chink_grammar)
```

Now, use the new chink parser to parse the tokens (sent3) and run the results.

```python
chink_parser.parse(sent_tokens3)
```

While comparing the syntax tree of chink parser with that of the original chunk, we can see that the token quickly (adverb) is chinked out of the chunk.

# Context Free Grammar  (CFG)


## CFG: Example
Let's understand the representation of Context Free Grammer(CFG) through an example.

* A sentence (S) can be represented as a Noun Phrase (NP) followed by a Verb Phrase (VP).
* Noun Phrase (NP) can de a Determiner Nominal(Det Nominal).
* A Nominal can be a Noun (Nominal --> Noun)
* VP represents the Verb Phrase (VP --> Verb)
* 'a' can be called as Determiner (Det --> a)
* Eg: Flight could be called as Noun (Noun --> Flight)




### Implementing CFG 
Consider the string below, where you have certain rules. We need to look at the given Context Free Grammer(CFG).

* A Sentence should have a noun phrase followed by a verb phrase. 
* A verb phrase is a verb followed by a noun. 
* A verb can either be saw or met. 
* Noun phrase can either be John or Jim, and 
* A noun can either be a dog or a cat.

```python
CFG_grammar = nltk.CFG.fromstring("""
S -> NP VP
VP -> V N
V -> "saw"|"met"
NP -> "John"|"Jim"
N -> "dog"|"cat"
""")
```

```python
import numpy as np
def generate_sample(grammar, prod, frags):        
    if prod in grammar._lhs_index: # Derivation
        derivations = grammar._lhs_index[prod]            
        derivation = np.random.choice(derivations)            
        for d in derivation._rhs:            
            generate_sample(grammar, d, frags)
    elif prod in grammar._rhs_index:
        # terminal
        frags.append(str(prod))
```

Check the possible list of sentences that can be generated using the rules, Use the join function to create the possible list of sentences.

```python
frags = []  
generate_sample(CFG_grammar, CFG_grammar.start(), frags)
print( ' '.join(frags) )
```

This dataset has tweets about six US Airlines along with their sentiments: positive, negative, and neutral. You are provided with this dataset named Tweets.csv. It has tweets in text column and sentiments in airline_sentiment column.

__Objective:__  Retrieve all tags starting from @ in the entire dataset and save in a file called References.txt. Extract all noun phrases from the dataset and save them in different lines in a file named Noun Phrases for <airline_sentiment> Review .txt (You can choose your own grammar for noun phrase). Here <airline_sentiment> will have three different values: positive, negative, and neutral. Hence, three files will be created.

```python
import os

pathlogs = 'C:\\Program Files\\gs\\gs9.54.0\\bin'
os.environ['PATH'] += os.pathsep+pathlogs

import pandas as pd
df = pd.read_csv('Tweets.csv')

#Extract the features text and airline sentiment
df = df[['text', 'airline_sentiment']]
```

```python
#iterate through the dataset using Regexp parser
#Find the relevant tweets
import re
r = re.compile(r"([@])(\w+)\b")

#import the itertools module, it returns efficient iterrators
AllReferences = map(lambda X: r.findall(X), df['text'])
```

```python
import itertools
AllUniqueReferencesCombined = set(list(itertools.chain.from_iterable(AllReferences)))
References = map(lambda X:X[0]+X[1], AllUniqueReferencesCombined)

file = open("References.txt",'a')
for each in References:
    file.write(each+"\n")
file.close()
```

```python
#Lets extact all noun phrases and save them in a file.
import nltk
from nltk import word_tokenize

def GetNounPhrases(S):
    try:
        sentences = nltk.sent_tokenize(S)
        sentences = [nltk.word_tokenize(sent) for sent in sentences]
        sentences = [nltk.pos_tag(sent) for sent in sentences]
    except:
         return []
    else:
        grammer = r"NP:{<DT><NN|NNS|NNP|NNPS>*<NN|NNS|NNP|NNP>}"
        cp = nltk.RegexpParser(grammer)
        noun_phrases_list = [[' '.join(leaf[0] for leaf in tree.leaves()) 
                              for tree in cp.parse(sent).subtrees() 
                              if tree.label() == 'NP'] 
                             for sent in sentences]
    return noun_phrases_list
```

```python
import itertools
for group, sub in df.groupby('airline_sentiment'):
    noun_phrases = map(lambda X: GetNounPhrases(X),sub['text'])
    noun_phrases = list(itertools.chain.from_iterable(noun_phrases))
    #Putting into list
    AllNounPhrases = set(list(itertools.chain.from_iterable(noun_phrases)))
    #Creating a file in the name of review.txt
    filename = "Noun Phrases for "+str(group)+" Review.txt"
    file=open(filename,'a',encoding='utf-8')
    for each in AllNounPhrases:
        file.write(each+"\n")
    file.close()
```

# Practice Project: Airline Sentiment
__Problem  Statement:__ ABC Company wants to perform some text analysis and make visualization for one its 
dataset. 

The dataset has been taken from Kaggle. (https://www.kaggle.com/crowdflower/twitter-airline-sentiment/home). 

This is a dataset having tweets about six US Airlines along with their sentiments: positive, negative, and neutral. You are provided with this dataset named Tweets.csv. It has tweets in text column and sentiments in airline_sentiment column.

__Objective:__  Extract all verb phrases from their dataset and save them in different lines in a file named Verb Phrases for <airline_sentiment> Review .txt (You can choose your own grammar for noun phrase). Here <airline_sentiment> will have three different values: positive , negative, and neutral. Hence, three files will be created. For each sentiment, make a well labeled pie chart showing the distribution of noun phrases and verb phrases of that sentiment from the data set. Use the files created above to get the frequencies.

```python
#Step 01
import pandas as pd
df=pd.read_csv('Tweets.csv')
df=df[['text','airline_sentiment']]
```

```python
#Step 02
import nltk
def GetVerbPhrases(s):
    try:
        sentences = nltk.sent_tokenize(s)
        sentences = [nltk.word_tokenize(sent) for sent in sentences]
        sentences = [nltk.pos_tag(sent) for sent in sentences]
    except:
        return []
    else:
        grammar=r"VP:{<VB|VBD|VBG|VBZ|VBP|VBN>*<VB|VBD|VBG|VBZ|VBP|VBN><RB|RBR>*<RB|RBR>}"
        cp = nltk.RegexpParser(grammar)
        verb_phrases_list = [[' '.join(leaf[0] for leaf in tree.leaves()) 
                              for tree in cp.parse(sent).subtrees() 
                              if tree.label()=='VP'] 
                             for sent in sentences]
    return verb_phrases_list
```

```python
#Step 03
import itertools
for group,sub in df.groupby('airline_sentiment'):
    verb_phrases=map(lambda x: GetVerbPhrases(x),sub['text'])
    verb_phrases=list(itertools.chain.from_iterable(verb_phrases))
    AllVerbPhrases=set(list(itertools.chain.from_iterable(verb_phrases)))
    filename="Verb Phrases for "+str(group)+" Review .txt"
    file=open(filename,'a',encoding='utf-8')
    for each in AllVerbPhrases:
        file.write(each+"\n")
    file.close()
```

```python
#Step 04
#Plotting a pie chart
def PlotPieChart(sentiment):
    noun_phrase_file="Noun Phrases for "+str(sentiment)+" Review .txt"
    verb_phrase_file="Verb Phrases for "+str(sentiment)+" Review .txt"
    noun_phrase_count=len(noun_phrase_file.split("\n"))
    verb_phrase_count=len(verb_phrase_file.split("\n"))
    counts=[noun_phrase_count,verb_phrase_count]
    labels=['Noun Phrases','Verb Phrases']
    
    import matplotlib.pyplot as plt
    %matplotlib inline
    plt.figure(figsize=(5,5))
    plt.pie(counts,labels=labels)
    plt.title("Phrases Distribution for "+str(sentiment)+" Review.")
    plt.show()

for each in df['airline_sentiment'].unique():
    PlotPieChart(each)
```
