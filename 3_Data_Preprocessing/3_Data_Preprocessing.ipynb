### Demo: Outlier and Missing Value Treatment
__Problem Statement__: Load the load_diabetes datasets internally from sklearn and check for any missing value or outlier data in the ‘data’ column. If any irregularities found treat them accordingly.

__Objective__: Perform missing value and outlier data treatment.

```python
#Step 1: Import the Libraries and Dataset
from sklearn.datasets import load_diabetes
dataset = load_diabetes()
dataset
```

```python
#Extract the data and the target independently out of it.
dataset.target
```

<font color='green'>__This gives us an array consisting data and target or label for training data. we are supposed to determine for another set of training data by training our model with respect to this target data.__</font>

```python
#Extract the column names out of the dataset.
dataset['feature_names']
```

<font color='green'>__It will be more convenient now to convert the dataset in the pandas dataframe for further operations. This dataframe ideally should include the column names, the training data and the target data from the dataset.__</font>

```python
#Step 2: Convert the dataset into DataFrame
#Import necessary Libraries
import pandas as pd
import numpy as np

#Use np.c_ to horizontally concentrate data and target values
df = pd.DataFrame(data=np.c_[dataset['data'],dataset['target']], columns=dataset['feature_names']+['target'])
df
```

```python
#Step 3: Check for any missing data or null values

df.isnull().any()
```

<font color='green'>__This shows no missing values in dataset.__</font>

```python
#Step 4: Check for Outliers in dataset
%matplotlib inline
import matplotlib.pyplot as plt

#Obtain box plot for each of the columns
for column in df:
    plt.figure()
    df.boxplot([column])
```

<font color='green'>__The boxplot for age variable is spread mostly between +0.02 to -0.02. Similarly Sex is between +0.04 to -0.04. However for the bmi variable, apart from the most of the data line between +0.03 and -0.03. Also there are few data points lying above 0.13. These data points are kind of isolated from all other data points. Hence these must be removed for successful modeling. We can detect outlier for the remaining varoables in simillar way. Now these outlier values can be treated accordingly with different outlier treatment techniques.__</font>
