## Paradigms of Recommendation Systems
Recommender systems reduce information overload by estimating relevance. As shown in the given image, a recommender system matches the relevence of the recommendation component with the recommendation list and based on the relevance, it decides the recommendation.


### Personalized Recommendation

The most relevant item is recommended based on the user profile and contextual parameters. Everyday massive amounts of information are generated and the existence of a large amount of information makes it difficult to find the required information.

Personalized recommendation is a process of alleviating the problem.



### Collaborative Filtering Recommendation

Let's understand how a recommendation system finds what's popular among your peers?

It uses the rating data to calculate the simillarity between items. It does not consider the schematic relationship between different items.


### Content-based Recommendation
It is use to display products simillar to the products we have liked before.

A content based recommender works witht the data provided by the user either explicitly through rating or implicitly by clicking on a link.

Based on a data provided by the user, It generates a user profile suggestions are given to user based on this.



The recommendation engine becomes more and more accurate as the user provides more inputs or takes actions on the recommendations.

### Knowledge-base Recommendation
Recommends the product based on users requirements. It uses knowledge about users products to make recommendations about which products meets the user's requirement.



###  Hybrid Recommendation

How a recommendation system can make combinations of various inputs & or composition of different mechanism?

It is a system that combines content based and collabrative filtering and takes advantage of both the representation of the content as well as the similarities among users.



It is also possible to use the recommendation of the two filtering techniques independently.

### Item-Based Nearest Neighbor
Because of the dynamic nature of the users, user based approach is often harder to scale as the items usually don't change much, thus Item based approach is preferred over user-based approach.

* Item based apprach are easy to scale, uses the similarity between items (and not users) to make predictions.
* Item based approach often can be computed offline and served without constant retraining.
* KNN is the perfect got to model to implement item based collbrative filtering. 
* It is also very good baseline for recommender system development.


From the example, In item based collaborative filtering we need to look for items that are like 'Item 5'
and then take Alice's rating from these items to predict the rating for 'Item 5'.



* The choice of the K nearest neighbours for the neighbouhood formation results in a trade off.
* When we use very small K, only a few candidates items can be recommended because there are not a lot of neighbours to support the predicitons.
* While the particularities of user preferences can be blunted due to the large neighbourhood size. So a very large 'K' value impacts precision.
* The optimum que also depends on data characterisitcs such as 'sparcity'. So K is set to be in the range of values from 1 to 100.

## Demo: User-Movie Recommendation Model
__Problem Statement:__ 

* Consider the ratings dataset below, containing the data on: UserID, MovieID, Rating and Timestamp. 
* Each line of this file represents one rating of one movie by one user, and has the following format:

        UserID::MovieID::Rating::Timestamp
        
* Ratings are made on a 5 star scale with half star increments. 
* UserID: represents ID of the user
* MovieID: represents ID of the movie
* Timestamp: represents seconds from midnight Coordinated Universal Time (UTC) of January 1, 1970.

__Objective:__ Predict a User-movie recommendation model

```python
#Step 1: Import Libraries and dataset
import pandas as pd
import numpy as np
df = pd.read_csv('Recommend.csv')
df.columns=['user_id','movie_id','ratings','timestamp']
df
```

```python
from sklearn.model_selection import train_test_split

#Step 2: Identify total no. of users and movies
#Creating a zero matrix with dimensions as n_users, n_movies
n_users = df.user_id.unique().shape[0]
n_movies = df.movie_id.unique().shape[0]

#Step 3: Split the data into training and testing data
train_data, test_data = train_test_split(df,test_size = 0.25) 

train_data_matrix = np.zeros((n_users,n_movies))
```

```python
#Step 4: populate the train and test metrics with random ratings 
#such that user_id index - movie_id index = given rating
for line in train_data.itertuples():
    train_data_matrix[line[1]-1, line[2]-1] = line[3]
    
train_data_matrix
```

```python
test_data_matrix = np.zeros((n_users,n_movies))
for line in test_data.itertuples():
    test_data_matrix[line[1]-1, line[2]-1] = line[3]
    
test_data_matrix
```

__Cosine Similarity__

Cosine similarity is a measure of similarity between two non-zero vectors of an inner product space that measures the cosine of the angle between them. The cosine of zero degree is '1' and it is less than '<1' for any angle in the interval (0*pi*r). where r is radius.

```python
#Step 5: Create cosine similarity metrices for user and movies
from sklearn.metrics import pairwise_distances
user_similarity = pairwise_distances(train_data_matrix, metric='cosine')
movie_similarity = pairwise_distances(train_data_matrix.T, metric='cosine')
mean_user_rating = train_data_matrix.mean(axis=1)[:,np.newaxis]
ratings_diff = train_data_matrix - mean_user_rating

#calculate user predicitons
user_pred = mean_user_rating + user_similarity.dot(ratings_diff)/np.array([np.abs(user_similarity).
                                                                           sum(axis=1)]).T
user_pred
```

Now, we have approximaely generated ratings for all movies.

## Movie-Movie Recommendation
__DESCRIPTION__

Consider the ratings dataset below containing data on UserID, MovieID, Rating, and Timestamp. Each line of this file represents one rating of one movie by one user and has the following format: 
* UserID::MovieID::Rating::Timestamp 

Ratings are made on a 5 star scale with half star increments. UserID: represents the ID of the user MovieID: represents the ID of the movie Timestamp: represents seconds from midnight Coordinated Universal Time (UTC) of January 1, 1970.

__Objective: Predict a movie-movie recommendation model.__


```python
#Step 1: Load the Ratings movie dataset into pandas with labels
df = pd.read_csv('Recommend.csv',names=['user_id', 'movie_id', 'rating', 'timestamp'])
df
```

```python
#Step 02: Create a train test split of 75/25 by declaring number of users and movies
n_users = df.user_id.unique().shape[0] 
n_movies = df.movie_id.unique().shape[0]
train_data, test_data = train_test_split(df, test_size=0.25)
```

```python
#Step 03: Populate the train matrix (user_id x  movie_id) with ratings 
#such that [user_id index, movie_id index] = given rating
train_data_matrix = np.zeros((n_users, n_movies))

for line in train_data.itertuples():
    #[user_id index, movie_id index] = given rating.
    train_data_matrix[line[1]-1, line[2]-1] = line[3] 
train_data_matrix
```

```python
#Step 04: Populate the test matrix (user_id x  movie_id) with ratings 
#such that [user_id index, movie_id index] = given rating
test_data_matrix = np.zeros((n_users, n_movies))

for line in test_data.itertuples():
    #[user_id index, movie_id index] = given rating.
    test_data_matrix[line[1]-1, line[2]-1] = line[3]
test_data_matrix
```

```python
#Step 05: Create cosine similarity matrices for movies and predict a movie-movie recommendation model
movie_similarity = pairwise_distances(train_data_matrix.T, metric='cosine')
movie_pred = train_data_matrix.dot(movie_similarity) / np.array([np.abs(movie_similarity).sum(axis=1)])
movie_pred
```
